# Quantitative Genetics- Sampling from multivariate Normal
We will be using the `Jupyter notebook` for some activities in this part of the QG 2024 course. Every notebook has an associated language called the `kernel`. We will be using Python 3 kernel from the IPython project.

We want to simulate random variables from normal distribution of given variables.

```python
import numpy as np
from scipy.stats import multivariate_normal
```

```python
# Define mean vector and covariance matrix
mu = np.array([-2, 3, 5])
sigma = np.array([
    [1.0, 0.3, -0.9],
    [0.3, 1.0, 0.0],
    [-0.9, 0.0, 1.0]
])
```

```python
# Print the covariance matrix
print("Covariance matrix:\n", sigma)
```

```python
# Generate 100 samples from the multivariate normal distribution
x = multivariate_normal.rvs(mean=mu, cov=sigma, size=100)
```

```python
# Print the first 10 values
print("First 10 samples:\n", x[:10])
```

Now let's plot these

```python
import matplotlib.pyplot as plt
```

```python
# Set up a 3x3 grid of subplots
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
# Generate the scatter plots
for i in range(3):
    for j in range(3):
        axes[i, j].scatter(x[:, i], x[:, j])
        axes[i, j].set_xlabel(f"Var{i+1}")
        axes[i, j].set_ylabel(f"Var{j+1}")

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```
